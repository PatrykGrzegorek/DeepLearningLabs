```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

import data
import fcann2
```

```python
np.random.seed(100)
tf.random.set_seed(100)

X, Y_ = data.sample_gmm_2d(4, 2, 30)

Y = data.myDummyDecision(X) > 0.5

bbox = (np.min(X, axis=0), np.max(X, axis=0))
data.graph_surface(data.myDummyDecision, bbox, offset=0)
data.graph_data(X, Y_, Y)

plt.show()
```

Exercise 2

```python
p2_x, p2_y = data.sample_gmm_2d(6, 2, 30)

p2_y_ = data.convert_to_one_hot(p2_y, 2)

p2_box = np.min(p2_x, axis=0), np.max(p2_x, axis=0)

p2_architecture = [2, 5, 2]
p2_n_epochs = 100000
p2_learning_rate = 0.05
p2_weight_decay = 1e-3

p2_model = fcann2.ANN(*p2_architecture)
p2_model.train(p2_x, p2_y_, p2_n_epochs, p2_learning_rate, p2_weight_decay)
```

```python
p2_fig = plt.figure(figsize=(16, 9))

data.graph_surface(p2_model.classify, p2_box, offset=0)
data.graph_data(p2_x, p2_y, p2_model.classify(p2_x))

plt.show()
```

Exercise 3


```python
## 1. definition of the computation graph
# data and parameters
tf.compat.v1.disable_eager_execution()
X  = tf.compat.v1.placeholder(tf.float32, [None])
Y_ = tf.compat.v1.placeholder(tf.float32, [None])
a = tf.Variable(0.0)
b = tf.Variable(0.0)
x_arr = [1,2]
y_arr = [3,5]

# affine regression model
Y = a * X + b

# quadratic loss
loss = (Y-Y_)**2

# optimization by gradient descent
trainer = tf.compat.v1.train.GradientDescentOptimizer(0.1)
train_op = trainer.minimize(loss)

## 2. parameter initialization
sess = tf.compat.v1.Session()
sess.run(tf.compat.v1.global_variables_initializer())

## 3. training
# let the games start!
for i in range(100):
    val_loss, _, val_a,val_b = sess.run(
      [loss, train_op, a,b], 
      feed_dict={X: x_arr, Y_: y_arr})
    print(i,val_loss, val_a,val_b)
```
